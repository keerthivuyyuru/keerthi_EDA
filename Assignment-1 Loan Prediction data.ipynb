- Relative frequency refers to the proportion of times a particular value occurs relative to the total number of values in a dataset.

**percentile-quantile**

- Percentiles:

     For example, the 50th percentile is the median, which divides the data into two equal parts.
   - np.percentile() function can be used to calculate percentiles. This function takes an array of values and a percentile as input and returns the value at that percentile.
- Quantiles:
    
    - Quantiles divide a dataset into q equal parts, where q is any number between 0 and 1. For example, the median is the 0.5 quantile.
    - where 0 represents the minimum value in the array and 1 represents the maximum value in the array.
    - np.quantile() function can be used to calculate quantiles. This function takes an array of values and a quantile as input and returns the value at that quantile.

**Outlier analysis**
- Outlier analysis the process of identifying data points that have extreme values compared to the rest of the distribution.


**Steps in outlier analysis**

-step1: Find the Q1,Q2 and Q3

   - np.percentile(column data,q)
    
-step2:Calculate the IQR

   - IQR= Q3-Q1
    
- Step-3: Calculate lower boundary and upper boundary

     - lb: Q1-1.5IQR
     - ub: Q3+1.5IQR
    
-Step-4: Find the Outliersdf

   - c1: column data < lb
   - c2: column data > ub
   - c: apply the main condition
   - main data[c]        

```python
# step1:

Q1=np.percentile(df['CoapplicantIncome'],25)
Q2=np.percentile(df['CoapplicantIncome'],50)
Q3=np.percentile(df['CoapplicantIncome'],75)

# step2:
IQR= Q3-Q1

# Step-3:
lb= Q1-1.5*IQR
ub= Q3+1.5*IQR
    
# Step-4: 
c1=df['CoapplicantIncome']>lb
c2=df['CoapplicantIncome']<ub
non_outliers_df=df[c1&c2]
non_outliers_df
```

**How to deal outliers**

***Drop the outliers*** 
 - it prevents skewing of the data
 - we can drop the outliers if outlier percentage < 2%
 -  But this is not recommended , we lost other columns data also

***Impute with Median values***
 - As we know that Median does not affect by outliers
 - So it is good prcatice we can impute outliers with Median value

***Cap with Q3 or Q1 value***
- If outliers are presnt less than lower bound then fill with Q1
- If outliers are more than upper bound then fill with Q3


```python
l=[]
median=df['CoapplicantIncome'].median()
for i in df['CoapplicantIncome']:
    if i<lb or i>ub:
        l.append(median)
    else:
        l.append(i)
print(len(l))
```

```python
# step1:

Q1=np.percentile(df['CoapplicantIncome'],25)
Q2=np.percentile(df['CoapplicantIncome'],50)
Q3=np.percentile(df['CoapplicantIncome'],75)

# step2:
IQR= Q3-Q1

# Step-3:
lb= Q1-1.5*IQR
ub= Q3+1.5*IQR

median=df['CoapplicantIncome'].median()

l=[]
for i in df['CoapplicantIncome']:
    if i<lb or i>ub:
        l.append(median)
    else:
        l.append(i)
df['CoapplicantIncome_new']=l
```

**np.where**

```python
# step1:

Q1=np.percentile(df['CoapplicantIncome'],25)
Q2=np.percentile(df['CoapplicantIncome'],50)
Q3=np.percentile(df['CoapplicantIncome'],75)

# step2:
IQR= Q3-Q1

# Step-3:
lb= Q1-1.5*IQR
ub= Q3+1.5*IQR

median=df['CoapplicantIncome'].median()

c1=df['CoapplicantIncome']<lb
c2=df['CoapplicantIncome']<lb
con=c1|c2

t=median
f=df['CoapplicantIncome']

df['CoapplicantIncome']=np.where(con,t,f)
plt.boxplot(df['CoapplicantIncome'])
plt.show()
```

```python
path = r"C:\Users\Keerthi\OneDrive\Documents\Full stack data science and ai\Loan Prediction\train_ctrUa4K - train_ctrUa4K.csv"
df=pd.read_csv(path)
df.head(5)
```

```python
# x-x_min/(x_max-x_min)
# step-1: Read the pawge column
# step-2: Find the min value of the pwage column
# step-2: Find the max value of the pwage column
# Step-3: nr= datacolumn-min value
# Step-4: dr= max_value-min_value
# Step-5: nr/dr
```

```python
# formula: x-x_min/(x_max-x_min)
# step-1: Read the ApplicantIncome column
ai=df['ApplicantIncome']

# step-2: Find the min and max value of the pwage column
ai_min=df['ApplicantIncome'].min()
ai_max=df['ApplicantIncome'].max()

# Step-3: nr= datacolumn-min value
nr=ai-ai_min

# Step-4: dr= max_value-min_value
dr=ai_max-ai_min

# Step-5: nr/dr
df['ApplicantIncome_N']=nr/dr
```
